# Covid-19 Vaccines Trendline

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

# yahoo_finance is used to fetch data 
import yfinance as yf
yf.pdr_override()
```

```python
# input
# Covid-19 Vaccines
title = "COVID-19"
symbols = ['BNTX', 'JNJ', 'MRNA', 'PFE']
start = '2020-01-01'
end = '2022-09-03'
```

```python
df = pd.DataFrame()
for s in symbols:
    df[s] = yf.download(s,start,end)['Adj Close']
```

```python
sns.set()
```

```python
df.head()
```

```python
df.tail()
```

```python
for s in df.columns:
    df[s].plot(figsize=(12,8))
    plt.title(s + "Price")
    plt.show()
```

```python
for s in df.columns:
    coefficients, residuals, _, _, _ = np.polyfit(range(len(df[s].index)), df[s],1, full=True)
    plt.figure(figsize=(12,8))
    plt.plot(df[s].values, label='Price')
    plt.plot([coefficients[0]*x + coefficients[1] for x in range(len(df))], label='Trendline')
    plt.title( s + ' Trendline')
    plt.legend(loc='best')
    plt.show()
```
